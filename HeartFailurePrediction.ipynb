#Death from heart failure predictions based on simple machine learning set on cloud

Aim of that project is to build machine learning oriented algorithm used for predicting death from heart failure, project uses paralell computing. That algorithm is based on dataset containing 12 features that can be used to predict mortality by heart failure. Dataset has been downloaded from  https://www.kaggle.com/andrewmvd/heart-failure-clinical-data

#1. Reading data

```
data = spark.read.csv('/FileStore/tables/heart_failure_clinical_records_dataset-1.csv',header=True, sep=',', inferSchema=True)
display(data) 
```

#2. Analyzing data

```
data.createOrReplaceTempView("HeartTable")
```

```
%sql
select age, DEATH_EVENT from HeartTable sort by age
```

```
%sql
select sex, DEATH_EVENT from HeartTable sort by sex
```

#3. Dataset development

```
from pyspark.ml.feature import VectorAssembler
train = VectorAssembler(inputCols = ["age", "anaemia", "creatinine_phosphokinase", "diabetes", "ejection_fraction", "high_blood_pressure", "platelets", "serum_creatinine", "serum_sodium", "sex", "smoking", "time", "DEATH_EVENT"], outputCol = "features").transform(data)
train1=train.withColumnRenamed("DEATH_EVENT", "label")
display(train1)
```

```
X,y = train1.randomSplit([0.8, 0.2])
```

#4. Training models

```
from pyspark.ml.classification import LogisticRegression
lr = LogisticRegression(maxIter=1, featuresCol = 'features', labelCol = 'label')
lrModel = lr.fit(X)
```

```
from pyspark.ml.classification import DecisionTreeClassifier
dt = DecisionTreeClassifier(featuresCol="features", labelCol="label")
dtModel = dt.fit(X)
```

```
from pyspark.ml.classification import RandomForestClassifier
rf = RandomForestClassifier(labelCol="label", featuresCol="features", numTrees=1)
rfModel = rf.fit(X)
```

#5. Test models and accuracy

```
lr_prediction = lrModel.transform(y)
dt_prediction = dtModel.transform(y)
rf_prediction = rfModel.transform(y)
```

```
from sklearn.metrics import confusion_matrix
from pyspark.ml.evaluation import BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator()
```

```
lr_acc = evaluator.evaluate(lr_prediction)
print("Logistic Regression prediction Accuracy: ", lr_acc)

y_pred=lr_prediction.select("prediction").collect()
y_orig=lr_prediction.select("label").collect()

cm = confusion_matrix(y_orig, y_pred)
print("Confusion Matrix:")
print(cm)

dt_acc = evaluator.evaluate(dt_prediction)
print("Decision Tree prediction Accuracy: ", dt_acc)

y_pred=dt_prediction.select("prediction").collect()
y_orig=dt_prediction.select("label").collect()

cm = confusion_matrix(y_orig, y_pred)
print("Confusion Matrix:")
print(cm)

rf_acc = evaluator.evaluate(rf_prediction)
print("Random Forest prediction Accuracy: ", rf_acc)

y_pred=rf_prediction.select("prediction").collect()
y_orig=rf_prediction.select("label").collect()

cm = confusion_matrix(y_orig, y_pred)
print("Confusion Matrix:")
print(cm)
```
